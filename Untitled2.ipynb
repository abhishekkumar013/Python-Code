```python
def read_score_file(filename):
    scores = {}
    with open(filename, 'r') as file:
        for line in file:
            try:
                data = line.strip().split('\t')  
                key = data[0].strip()
                value = float(data[1].strip())
                scores[key] = value
            except IndexError:
                print("Error occurred with line:", line)
    return scores


def merge_scores(score_files, assignment_file):
    assignment_scores = read_score_file(assignment_file)
    merged_scores = {}
    for file in score_files:
        scores = read_score_file(file)
        for key, value in scores.items():
            if key in assignment_scores:
                name = key.split(' ')[1]  
                if name in merged_scores:
                    merged_scores[name].append(value)
                else:
                    merged_scores[name] = [value]

    
    with open("final_scores.txt", "w") as final_file:
        for key, value in merged_scores.items():
            avg_score = sum(value) / len(value)
            final_file.write(f"{key},{avg_score}\n")

if __name__ == "__main__":
    score_files = ["RollNo_Score_Quiz_1.txt", "RollNo_Score_Quiz_2.txt", "Name_Viva.txt"]
    assignment_file = "RollNo_Name_Assignment.txt"
    merge_scores(score_files, assignment_file)
```
